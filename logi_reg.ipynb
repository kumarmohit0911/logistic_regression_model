```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
#creating a dataset
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
X
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
#model training
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# accuracy checking
score= accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

##  Hyperparameter Tuning and Cross Validation

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg','lbfgs','liblinear','sag','saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

# GridSearchCV

```python
#GridSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

## Logistic Regression For Multiclass Classification Problem

```python
# creating dataset 
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000,n_features=10,n_informative=3,n_classes=3,random_state=42)
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
pd.DataFrame(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression(multi_class='ovr')
```

```python
logistic
```

```python
logistic.fit(X_train,y_train)
y_pred=logistic.predict(X_test)
```

```python
y_pred
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

## Hyperparameter tuning and Cross validation

```python
multimodel=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg','lbfgs','liblinear','sag','saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
#GridSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
grid=GridSearchCV(estimator=multimodel,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred=grid.predict(X_test)
```

```python
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```
